# CSV Files Line Comparison
This notebook will:
1. Find all CSV files in specified directories
2. Read their contents
3. Count how many times each line appears across different files
4. Display the results

```python
import os
import pandas as pd
from pathlib import Path
from collections import defaultdict

def find_csv_files(root_dir='.'):
    """Find all CSV files in the given directory and its subdirectories"""
    csv_files = []
    for path in Path(root_dir).rglob('*.csv'):
        csv_files.append(str(path))
    return csv_files

# Find all CSV files
csv_files = find_csv_files("exports/wos")
print(f"Found {len(csv_files)} CSV files:")
for file in csv_files:
    print(f"- {file}")
```

```python
def read_csv_files(directory='.'):
    csv_files = [f for f in Path(directory).rglob('*.csv')]
    return csv_files

def process_files(csv_files):
    dfs = []  # List of dataframes (one per ffile/query)
    for file in csv_files:
        try:
            df = pd.read_csv(file, sep=';')
            dfs.append(df)
        except Exception as e:
            print(f"Error processing {file}: {str(e)}")
    return dfs

# Read and process all CSV files
csv_files = read_csv_files(directory='./wos')
dfs = process_files(csv_files)
```

```python
# Merge all dataframes
merged_df = pd.concat(dfs, ignore_index=True)

# Count duplicates
duplicate_counts = merged_df.duplicated().sum()
print(f"Total number of duplicate rows: {duplicate_counts}")

key_columns = ['Author Full Names', 'Article Title', 'Source Title']
duplicate_counts = merged_df.duplicated(subset=key_columns).sum()
print(f"Total number of duplicate rows: {duplicate_counts}")

# To see which rows are duplicated and their counts
duplicates_with_counts = merged_df.value_counts(subset=key_columns, dropna=False).reset_index()
duplicates_with_counts.columns = [*key_columns, 'count']
print("\nDuplicate rows and their counts:")
print(duplicates_with_counts[duplicates_with_counts['count'] > 1])

```
